Tuple adalah jenis dari struktur data yang dapat diubah elementnya. Umumnya tuple digunakan untu data yang bersifat sekali tulis, dan dapat dieksekusi lebih cepat. Tuple didefinisikan dengan kurung dan element yang dipisahkan dengan koma.

```python
#contoh tuple 
T1 = (1, "dua", True, 1+2j)
print("Ini adalah Tuple", T1)
```

```python
#Menampilkan tipe struktu data
type(T1)
```

**Latihan**

```python
#Buatlah tupleu sendiri
tup = ('matrix', 225, 0, 'fondasi', True)
print(tup, type(tup))
```

**Sifat-sifat Tuple**

```python
#Terurut/Orered
T1[0:3]
```

```python
#Tidak Dapat Diganti/Immutable
T1[2] = 'Manusia'
```

```python
#Dapat Diindeks/Slicing
print("elemen kedua dari tuple T1:", T1[1])
```

```python
#Dapat Diijinkan Duplikasi
T2 = (3, 2.4, True, 2.4)
T2
```

**Menggabungkan Tuple**

```python
#Menggabungkan Tuple
T3 = T1 + T2
T3
```

**Latihan 2**

```python
#Buatlah tuple kedua lalu gabungkan 2 tuple yang telah kamu buat
tup2 = (26, 'bird', 11.02, True, 26)
tup3 = tup + tup2
tup3
```
