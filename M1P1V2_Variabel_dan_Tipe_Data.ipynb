# Variabel

<strong>Variables</strong>: Lokasi penyimpanan yang dapat digunakan untuk menampung sebuah data atau informasi.



Syarat-syarat penamaan variabel 
1. Karakter-karakter yang digunakan untuk penamaan variable adalah alphabet, angka dan underscore. 
2. Karakter pertama variable harus berupa huruf atau garis bawah/underscore, dan tidak bisa berupa angka. 
3. Nama sebuah variabel tidak bisa menggunakan keyword atau reserved words dari bahasa python seperi for, if dan lainnya. 
4. Karakter pada nama variabel bersifat sensitif (case-sensitif).

```python
#membuat variabel 

kalimat = 'ini adalah variabel' 
bilangan1 = 4 
bilangan2 = 2.5 
bilangan3 = True 
bilangan4 = 1j + 4
```

```python
# menampilkan variabel 

print(kalimat) 
print(bilangan1) 
print(bilangan2)
```

Latihan 1 dan Latihan 2

```python
# buatlah 5 variabel kalian sendiri dengan 5 macam tipe data 
# lalu tampil kan variabel menggunakan fungsi print

Data_1 = 6
print(Data_1, type(Data_1))
print('-'*20)

Data_2 = 15.06
print(Data_2, type(Data_2))
print('-'*20)

Data_3 = 'Matematika 2021'
print(Data_3, type(Data_3))
print('-'*20)

Data_4 = ('Matematika', 'UAD', 2021)
print(Data_4, type(Data_4))
print('-'*20)

Data_5 = {'tinggi':158,'massa':50}
print(Data_5, type(Data_5))
print(Data_5['tinggi'])
print('-'*20)

Data_6 = False
print(Data_6, type(Data_6))
print('-'*20)

Data_7 = 8j + 3
print(Data_7, type(Data_7))
print('-'*20)
```

# Tipe Data

5 tipe data yaitu 
1. Integer atau bilangan bulat 
2. Float atau bilangan desimal 
3. String atau kalimat 
4. Boolean yang terdiri dari True dan False 
5. Complex yang merupakan bilangan kompleks

# Konversi Tipe Data

Syntax Konversi tipe data 
1. Integer = int() 
2. Float = float() 
3. String= str() 
4. Boolean = bool() 
5. Complex = Complex()

```python
# konversi tipe data integer 
print(int(bilangan2)) 
print(int(bilangan3))
```

```python
# konversi tipe data float 
print(float(bilangan1)) 
print(float(bilangan3))
```

```python
# konversi tipe data string 
print(str(bilangan1)) 
print(str(bilangan2)) 
print(str(bilangan3))
```

```python
#konversi tipe data bool 
print(bool(bilangan1)) 
print(bool(bilangan2))
```

```python
# konversi tipe data complex 
print(complex(bilangan1)) 
print(complex(bilangan2)) 
print(complex(bilangan3))
```
