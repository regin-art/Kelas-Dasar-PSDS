list adalah jenis kumpulan data terurut (ordered sequence), dan merupakan salahsatu variabel yang sering digunakan pada Python. Serupa, namun tak sama dengan array pada bahasa pemrograman lainnya. Bedanya, element list pada Python tidak harus memiliki type dataa yang sama. Mendeklarasikan list cukup mudah dengan kurung siku dan element yang dipisahkan dengan koma.

Setiap data di dalamnya dapat diakses dengan indeks yang dimulai dari 0. 

```python
#contoh List 
L1 = [1, 4, 2, 13]

print(type(L1))
```
