Python pada umumnya terdiri dari

<strong>2. Varibles</strong>

```python
bilangan1 = 5 
bilangan2 = 10 
kalimat = "Pejuang Data 3"
```

<strong>3. Literals</strong>: Simbol-simbol yang dapat kita gunakan untuk mengisi suatu variabel.

<strong>4. Operators.</strong> Simbol-Simbol yang dapat digunakan untuk mengubah nilai dari satu variabel dengan melibatkan satu atau lebih variabel dan literal.

```python
#penjumlahan
jumlah = bilangan1 + bilangan2
print(jumlah)
#pengurangan
kurang = bilangan1 - bilangan2
print(kurang)
#perkalian
kali = bilangan2 * (bilangan1 - 3)
print(kali)
#pembagian
bagi = bilangan2/bilangan2 #dividing any two integers produces a float. 
print(bagi)
#pangkat
pangkat = bilangan1**2
print(pangkat)
```

<strong>5. Reserved Words.</strong> Kumpulian kata-Kata yang memiliki makna Khusus dalam bahasa pemrograman Python. Kata False, return, dan for merupakan contoh dari reserved words.

```python
#True False
print(bilangan1<bilangan2)
print(bilangan1==bilangan2)
print(bilangan1>bilangan2)
```

```python
#for
for a in range (6):      #mulai dari 0
   print(a)

for i in range (10):     #in range (10) = 0,1,2,...9
    if i==3:             #0,1,2
       continue          
    elif i==5:           #i = 4
       break
    print(i**2)
```

<strong>6. Whitespace</strong>: Pada bahasa Python, spasi dan tab memiliki makna khusus untuk menandai serangkaian blok dalam kode Python.

<strong>7. Comments</strong>: Comments merupakan sekumpulan teks yang dituliskan di dalam sebuah program yang tidak akan mempengaruhi hasil dari sebuan program.

Terdapat dua jenis comments di dalam Python, yaitu: 



<strong>a. single line comment</strong> (comments dalam satu baris) seperti pada contoh berikut:

```python
#Ini adalah Comment 
print ("belajar Python")
```

<strong>b. multi line comment</strong> (comments dalam beberapa baris) yang dapat dituliskan seperti pada contoh berikut:

```python
'''
Ini adalah komentar panjang
Bisa digunakan untuk memberikan penjelasan 
panjang pada Sell Kode Python,
Misal dokumentasi program, spesifikasi 
input output atau lainnya 
'''

print("Belajar Python menyenangkan")
```
